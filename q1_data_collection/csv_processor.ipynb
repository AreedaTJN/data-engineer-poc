```python
# Import necessary libraries
import pandas as pd
```

```python
# CSV file path
tci_path = '../data/API_TCI_OUTPUT.csv'
scopus_path = '../data/API_SCOPUS_OUTPUT.csv'
wos_path = '../data/API_ISI_OUTPUT.csv'

# Read CSV files
tci_data = pd.read_csv(tci_path)
scopus_data = pd.read_csv(scopus_path)
wos_data = pd.read_csv(wos_path, encoding='cp1252')
```

```python
# Select specific columns from each dataset
tci_cols = ['ARTICLE_NAME', 'AUTHOR', 'OUTPUT_YEAR', 'JOURNAL', 'VOLUME', 'ISSUE', 'PAGE_RANGE']
scopus_cols = ['TITLE', 'AUTHOR', 'OUTPUT_YEAR', 'JOURNAL', 'VOLUME', 'ISSUE', 'PAGE_RANGE', 'DOC_TYPE']
wos_cols = ['ARTICLE_TITLE', 'AUTHOR', 'OUTPUT_YEAR', 'JOURNAL', 'VOLUME', 'ISSUE', 'PAGE_RANGE', 'DOC_TYPE']

tci_selected = tci_data[tci_cols]
scopus_selected = scopus_data[scopus_cols]
wos_selected = wos_data[wos_cols]
```

```python
# Display the first few rows of the selected data
# tci_selected.head()
# scopus_selected.head()
wos_selected.head()
```

```python
# Add a 'SOURCE' column to each DataFrame
tci_selected = tci_selected.copy()
scopus_selected = scopus_selected.copy()
wos_selected = wos_selected.copy()
tci_selected['SOURCE'] = 'tci'
scopus_selected['SOURCE'] = 'scopus'
wos_selected['SOURCE'] = 'wos'
```

```python
# Export data to CSV files
tci_selected.to_csv('../data/tci_selected.csv', index=False)
scopus_selected.to_csv('../data/scopus_selected.csv', index=False)
wos_selected.to_csv('../data/wos_selected.csv', index=False)
```
